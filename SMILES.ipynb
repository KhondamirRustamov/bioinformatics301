```python
#@title Input molecule in smiles portrait and draw it

from rdkit import Chem
from rdkit.Chem import Draw

smiles = 'c1ccccc1C=CC(C=O)CO' #@param {type:"string"}

m = Chem.MolFromSmiles(smiles)
img = Draw.MolToImage(m)
img
```
